## Part 2: Exploratory Data Analysis (EDA)

```python
##Now we are asked to provide a summary statistics for each attribute so we do df.describe to check the summary statistics
df.describe()
```

```python
#Now we are asked to provide appropriate graphical analysis for each attribute so we import matplotlib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# I dropped the Unnamed column because it doesn’t give us useful info so it wasn't needed.
df_filtered = df.drop(columns=['Unnamed: 0'])
```

```python
#Pairplot shows the relationship betweeen each variable if there is one
#Used sns(seaborn)
# We're specifically selecting 'x', 'y', 'z', and table because they're the physical dimensions of diamonds.
sns.pairplot(df[['x', 'y', 'z', 'table']])
plt.suptitle('Pairwise Relationships: x, y, z, and table', y=1.02)
# The "y=1.02" just moves the title slightly above the plot so it doesn't overlap
plt.show()
```

```python
#Shows the distrubution of the variables: carat, depth, table, etc.
numeric_columns = df.select_dtypes(include='number').columns

# Loop through each numeric column to create a histogram
for column in numeric_columns:

#Used AI to help with the making of this graph.
#The kde = true creates a histogram with a smooth curve to show the distribution shape
    sns.histplot(df[column], kde=True)
    plt.title(f'{column} Distribution')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

### Part 2: Analysis Summary

After creating the graphs for the data, i found that in the pairplots, we're looking for graphs that has dots scattered on a straight line which tells us that the variables has a strong relationship to them. After printing the graph, I see that X(length) has no relationship with the table(Width).

The length, width, and depth of the diamonds are closely related. That makes sense because bigger diamonds are larger in all dimensions.

The table, which is the top flat part of the diamond, doesn’t have a strong relationship with the other sizes. This means that even diamonds with similar size can have slightly different table widths.

There are also a few dots far away from the main groups, which could be mistakes in the data or very unusual diamonds.

Basically, Think of it like comparing the height, width, and depth of different boxes — usually, if one is big, the others are too. But the "table" is like the design of the box’s top — it doesn’t always scale the same way.

#### **Now in the histogram**:



**Carat Distribution**: We can see that most diamonds in the dataset are small, under 1 carat. The histogram peaks in that range, meaning small diamonds are the most common. As the carat size increases, the number of diamonds drops off quickly. This shows that bigger diamonds are much rarer.

**Z Distribution (Depth in mm)**: Most diamonds have a depth between 2 and 5 mm. There are a few extreme values which may be errors or outliers. This again shows a consistent size pattern for most diamonds.

**Y Distribution (Width in mm)**: Like depth, most diamonds fall between 3 and 8 mm in width. A few are much wider, which could be that the daimonds are rare

**X Distribution (Length in mm)**: Length also follows the same trend, most diamonds are 4 to 7 mm long. It's rare to find ones longer than that.

**Price Distribution**: Most diamonds cost between 500 and 4,000. As the price increases, the number of diamonds drops sharply. This reflects the pricing of larger or better quality diamonds.

**Table Distribution**: The table is the flat top of the diamond. Most diamonds have a table between 55% and 60%, which shows that there's a preferred standard in cutting for best light reflection.

```python
#6  Using both Matplolib and Seaborn, make boxplots of the diamond price for each distinct cut value
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='cut', y='price')

plt.title('Price of Diamond by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
#First we establish a size with the figure function. 10 inches wide and 6 inches tall
#We then create a boxplot using sns (seaborn) package.
#Our data parameter is the data set
#our X is going to be "cut" because it is catagorical.
#Our Y is price because it is a continous variable.

```

The conclusion that I can draw from the graph is that premium cut daimonds have the highest mean price.

```python
#7. Using both Matplolib and Seaborn, make a scatter plot of price vs. carat. What can we say about the relationship between those two attributes
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='carat', y='price')

plt.title('Diamond Price vs Carat')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
#using sns to make a scatter plot

```

This scatter plot shows the relationship between a diamond’s weight in carats and its price

As we can see, there’s a clear upward trend. Which looks like the more carats a diamond has, the higher the price.
